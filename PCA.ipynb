```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.decomposition import PCA
```

```python
data = np.loadtxt('./datasets/data.txt')
covar = np.cov(data.T)
eig_vectors= np.linalg.eig(covar)[1]
u_red= eig_vectors[:,:2]
reduced=np.dot(data,u_red)
plt.scatter(reduced[:,0], reduced[:,1]) 
plt.show()
```

```python

model = PCA(n_components=99)
# model.fit(data[:,:-1])
print(model.fit(data[:,:-1]))
print(model.explained_variance_)

# data = model.fit_transform(data[:,:-1],data[:,-1])

# plt.scatter(data[:,0],data[:,1])
# plt.show()
```
